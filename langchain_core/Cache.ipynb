# Caching

```python
import time
from langchain_core.caches import InMemoryCache
from langchain_core.globals import set_llm_cache
from langchain_ollama import ChatOllama

# Enable the Global In-Memory Cache
set_llm_cache(InMemoryCache())

#  Set up the LLM
llm = ChatOllama(model="deepseek-r1:8b")


def test_cache(prompt):
    start = time.time()
    # The LLM call
    result = llm.invoke(prompt).content
    duration = time.time() - start

    print(f"Prompt: '{prompt}'")
    print(f"Output: {result}")
    print(f"Time:   {duration:.6f} seconds")
    print("-" * 40)


# --- EXECUTION ---

print("--- Round 1: Initial Call (Populating Cache) ---")
test_cache("Optimize my LakeFlow pipeline give me one line answer")

print("--- Round 2: Exact Same Call (Hit Cache) ---")
# This should be near 0 seconds and return the SAME response as Round 1
test_cache("Optimize my LakeFlow pipeline give me one line answer")

print("--- Round 3: Different Prompt (Cache Miss) ---")
test_cache("Check Auto Loader status Give me one line answer")
```
