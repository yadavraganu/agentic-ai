# Context
It refers to static information injected into an agent at runtime — such as user IDs, database connections, or configuration details — so that tools and middleware can access them without relying on global state. This makes agents more flexible, reusable, and testable.
- Holds static information like user ID, database connections, or configuration.
- Enables dependency injection instead of hardcoding values.
- Defined via context_schema when creating an agent.

```python
from dataclasses import dataclass
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

llm = ChatOllama(model="gemma4:e4b")


@dataclass
class Context:
    user_name: str


def get_user_name(context: Context) -> str:
    return context.user_name


agent = create_agent(llm, context_schema=Context)

response = agent.invoke(
    {"messages": [{"role": "user", "content": "What's my name?"}]},
    context=Context(user_name="John Smith"),
)

utils.response_printer(response)
```
